```python
pattern_url = re.compile(r'http\S+|www.\S+')                 # To identify URLs
pattern_email = re.compile(r'\S+@\S+')                       # To identify emails
string_without_urls = re.sub(pattern_url, 'URL', data)
f_string = re.sub(pattern_email, 'EMAIL', string_without_urls)

print(f_string)
print(string_without_urls)
```

```python
# Getting the number of lines
pattern = r'Lines:\s*(\d+)'
match = re.search(pattern, f_string)
num=[]
if match:
    num = int(match.group(1))
    print(num)
```

```python
f_str = f_string.split('\n')
if f_str[-1] == "":
    del f_str[-1]
print(f_str)
```

```python
lst = f_str[-num:]
print(lst)
```

```python
# Delete lines that contain the #reply# pattern
lst = [line.replace('#reply#', '') for line in lst]
print(lst)
```

```python
ann = []
for line in lst:
    if "#sig#" in line:
        ann.append(1)
    else:
        ann.append(0)
        
print(ann)
```
