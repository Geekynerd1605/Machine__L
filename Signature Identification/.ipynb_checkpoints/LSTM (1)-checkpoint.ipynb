```python
#create a list of all the entities that can be retrieved in the text using regex (Email, Tel, Web)

# Function to retrieve different entities in text using regex 

def find_regex_pattern(text, type_, pattern):

    list_result = []
    results = regex.finditer(pattern, text, flags=regex.IGNORECASE)
    for match in results:
        value = match.string[match.start(): match.end()].strip()
        list_result.append([type_,
                            value,
                            match.start(),
                            match.end(),
                            1])
    return list_result


dict_pattern = dict(EMAIL=r'[\p{L}\p{M}\-\d._]{1,}@[\p{L}\p{M}\d\-_]{1,}(\.[\p{L}\p{M}]{1,}){1,}')

list_result = []
for type_, pattern in dict_pattern.items():
    result = find_regex_pattern(text, type_, pattern)
    list_result += result
list_result
```

```python
# Then we create a list of all the entities that can be retrieved in the text (PERSON, ORG, LOC)

from nltk.tokenize import word_tokenize

tokenizer = word_tokenize(text)

NER = spacy.load("en_core_web_lg")
results = NER(text)

for result in results.ents:
   # print(result.text, result.label_)
    if result.text != "" and result.label_ in ["PERSON", "LOC", "ORG"]:
        list_result.append([result.label_, 
                            result.text,
                            result.start_char, 
                            result.end_char])
print(list_result)
```

```python
# We convert features list to dataframe
df_ner = pd.DataFrame(list_result, columns=["entity", "value", "start", "end", "score"])
df_ner
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras import optimizers
from keras.layers import Dense, LSTM
from keras.models import Sequential
from scipy.spatial import distance
from spacy import displacy
import seaborn as sns
import pandas as pd
import numpy as np
import spacy
import regex
import re
```

```python
NER = spacy.load("en_core_web_lg")
```

```python
embedder_model = SentenceTransformer("distiluse-base-multilingual-cased-v1")
embedding_thanks = embedder_model.encode('thanks')
```

```python
email = pd.read_csv("new.csv")
```

```python
list_line_number = []
list_line = []
list_word_count = []
list_per = []
list_org = []
list_loc = []
list_email = []
similarity_with_thanks = []
list_position_line = []
list_special_char = []
list_tel = []
```

```python
for index in range(email.shape[0]):
    message = email.iloc[index]['message']
    #print(message)
    #list_email.append(message.split('From: ')[1].split('\n')[0])
    text = message.split('X-FileName: ')[1].lstrip('\n')
    text = text[text.find('\n'):]
    
    lines = text.split('\n')    
    line_index = 0


    for line in lines:
        line = line.strip(" ")
        if len(line) == 0:
            continue
           
        #print(line)
        
        list_line_number.append(line_index)
        line_index += 1
        list_line.append(line)
        
        
        pattern = r"\b(\d{3}[-.\s]??\d{3}[-.\s]??\d{4}|\(\d{3}\)\s*\d{3}[-.\s]??\d{4})\b"
        matches = re.search(pattern, line)
        if matches is not None and matches[0]==line:
            list_tel.append(1)
        list_tel.append(0) 
        
        
        list_word_count.append(len(line.split()))
        list_position_line.append(line_index)
        
        results= NER(line)
   
        found=0
        for word in results.ents:
            if word.text != "" and word.label_ in ["PERSON"]:
                list_per.append(1)
                found=1
                break

        if found==0:
            list_per.append(0)
            
        found=0
        for word in results.ents:
            if word.text != "" and word.label_ in ["LOC"]:
                list_loc.append(1)
                found=1
                break

        if found==0:
            list_loc.append(0)

        found=0
        for word in results.ents:
            if word.text != "" and word.label_ in ["ORG"]:
                list_org.append(1)
                found=1
                break

        if found==0:
            list_org.append(0)


        special_char_count = len(regex.findall(r"[^\p{L}0-9 .,\n]", line))
        list_special_char.append(special_char_count)
       
        #doc_line = nlp(line)
        embedding_line = embedder_model.encode(line)
        dist = distance.cosine(embedding_thanks, embedding_line)
        #similarity_with_thanks.append(doc_thanks.similarity(doc_line))
        similarity_with_thanks.append(1 / (dist + 0.01))
```

```python
# scale all features by using MinMaxScaler for all features
# Select the columns to be scaled
cols_to_scale = ['PER', 'ORG', 'LOC', 'TEL', 'word_count', 'inv_distance_to_thanks', 'position_line', 'special_characters_count']

# Initialize the scaler
scaler = MinMaxScaler()
scaler.fit(df[cols_to_scale])
df[cols_to_scale] = scaler.transform(df[cols_to_scale])

df
```

```python
x = df[cols_to_scale].to_numpy()
x = np.expand_dims(x, axis=0)
x.shape
```

```python
y = np.array(ann)
y = np.expand_dims(y, axis=0)
y.shape
```

```python
import keras
```

```python
model = keras.Sequential()

model.add(keras.layers.Bidirectional(layer=keras.layers.LSTM(units=40,return_sequences=True,dropout=0.15)))
model.add(keras.layers.Dense(1, activation='sigmoid'))
opt = keras.optimizers.Adam()
model.compile(loss="binary_crossentropy",optimizer=opt,metrics=[keras.metrics.Precision(name="precision"),
                                                 keras.metrics.Recall(name="recall")])
model.build(x.shape)
model.summary()
```

```python
history = model.fit(x,y,epochs=200, verbose=1)
```

```python
y_predict = model.predict(x)>0.5
y_predict
```

```python
import numpy as np
from sklearn.metrics import precision_recall_fscore_support
```

```python
# Convert the label arrays to binary matrices
y_true_binary = np.argmax(y, axis=1)
y_pred_binary = np.argmax(y_predict, axis=1)
```

```python
# Calculate precision, recall, and F1 score for each class
precision, recall, f1_score, _ = precision_recall_fscore_support(y_true_binary, y_pred_binary, average=None)
```

```python
# Calculate weighted F1 score
weighted_f1_score = f1_score.mean()

# Print the results
print('Precision:', precision)
print('Recall:', recall)
print('F1 score:', f1_score)
print('Weighted F1 score:', weighted_f1_score)
```
